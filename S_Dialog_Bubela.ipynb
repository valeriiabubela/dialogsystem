```python
def dialogmanager():
    while input_s != "fertig":
    input_s = do_input()
    semantics = semantic(input_s)
    results = data(semantics)
    print(output(results))
```

```python
def do_input():
    return "to be implemented"
```

```python
def output(results):
    return "das Wetter ist {}".format(results)
```

```python
def semantic(input_s):
    if input_s[0] == "a":
        return "x"
    else:
        return "y"    
```
